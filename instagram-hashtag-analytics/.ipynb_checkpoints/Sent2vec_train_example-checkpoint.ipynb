```python
# load
class HashtagDoc2Vec:
    def __init__(self, fname):
        self.fname = fname
        if not os.path.exists(fname):
            print('File not found: %s' % fname)
    
    def __iter__(self):
        with open(self.fname, 'r', encoding='utf-8') as f:                    
            for row in f :    
                cm_id, pm_id, pm_code, pm_tag, pm_view, pm_like, pm_comment = row.split('\t')
                tagged_doc = TaggedDocument(
                    words = pm_tag.split(),
                    tags = [pm_code]
                )
                yield tagged_doc
```

```python
openPath = '../data/instagram/post/sent2vec_ta_hashtag_sampled.txt'
sent2vec_corpus = HashtagDoc2Vec(openPath)
```

```python
 for i, doc in enumerate(sent2vec_corpus):
    if i > 3: break
    print(doc)
```

```python
%%time
TRAIN_DOC2VEC = True
doc2vec_fname = '../data/instagram/post/hashtag_sent2vec_model_filtered.pkl'
'''
학습을시킬때 sentence를 고려한 설정값이 있을것... 
fasttext로 학습시키는것같기도 하고...
확인필요
'''
if TRAIN_DOC2VEC : 
    doc2vec_model = Doc2Vec(
        sent2vec_corpus,
        dm=1,
        vector_size=200,
        alpha=0.025,
        min_alpha=0.015,
        window=5,
        min_count=10,
        workers=8,
        negative=5
    )
    with open(doc2vec_fname, 'wb') as f:
        pickle.dump(doc2vec_model, f)
else :
    with open(doc2vec_fname, 'rb') as f:
        doc2vec_model = pickle.load(f)
```
