### pymagnitude library
- pip3 install pymagnitude 설치
- 벡터연산을 할때 쓰는 라이브러리인것 같다..
- model.txt 파일을 .magnitude 포맷으로 바꾸어 사용
- gensim(26Mb)보다 큰 패키지이다(110Mb)...

```python
from pymagnitude import Magnitude
mpath = '../data/ta_hashtag/vectors.magnitude'
vectors = Magnitude(mpath)
```

```python
print(vectors.query('일상'))
```

```python
print(vectors.most_similar(positive=['여행','맛집'], negative=['호텔'], topn=20))
```
